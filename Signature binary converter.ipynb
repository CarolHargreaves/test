```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show(image, title = 'Image', figsize = (5, 5)):
    plt.figure(figsize = figsize)
    plt.imshow(image, cmap = 'gray')
    plt.title(title)
    plt.axis('off')
```

```python
raw = cv.imread('./data/Signature.jpeg')
show(raw)
```

```python
gray = cv.cvtColor(raw, cv.COLOR_BGR2GRAY)
show(gray)
```

```python
binary = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 501, 5)
show(binary)
```

```python
cv.imwrite('./data/sig1.jpg', binary
          )
```
